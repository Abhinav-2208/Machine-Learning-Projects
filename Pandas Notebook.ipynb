### Agregate method --> passing dictionary

different columns and genres, we can apply different type of aggregation function like min/max/sum/mean etc.

```python
genres.agg({
    'Runtime':'mean',
    'IMDB_Rating':'mean',
    'No_of_Votes':'sum',
    'Gross':'sum',
    'Metascore':'min'
})
```
